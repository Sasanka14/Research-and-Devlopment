```python
import os
import shutil
import kagglehub

# ---------------- CONFIG ----------------
DATASET_HANDLE = "farzadnekouei/top-view-vehicle-detection-image-dataset"
TARGET_DIR = "/Users/zoro/Developer/Projects/traffic - congestion/asset"
# ---------------------------------------

print("Downloading dataset from Kaggle...")
cache_path = kagglehub.dataset_download(DATASET_HANDLE)
print("KaggleHub cache path:", cache_path)

# Ensure target directory exists
os.makedirs(TARGET_DIR, exist_ok=True)

# Copy dataset contents into your asset directory
print("Syncing dataset into project asset directory...")
for item in os.listdir(cache_path):
    src = os.path.join(cache_path, item)
    dst = os.path.join(TARGET_DIR, item)

    if os.path.isdir(src):
        shutil.copytree(src, dst, dirs_exist_ok=True)
    else:
        shutil.copy2(src, dst)

print("\n✅ Dataset successfully added to:")
print(TARGET_DIR)

print("\n📁 Asset folder structure:")
for root, dirs, files in os.walk(TARGET_DIR):
    level = root.replace(TARGET_DIR, "").count(os.sep)
    indent = " " * 4 * level
    print(f"{indent}{os.path.basename(root)}/")
    for f in files:
        print(f"{indent}    {f}")
```
